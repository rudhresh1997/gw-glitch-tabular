```python
# Slide 3 — Class Distribution of Glitch Types (log scale)
import pandas as pd
import matplotlib.pyplot as plt

# Load your cleaned dataset
df = pd.read_csv("original_data_cleaned.csv")

# Count occurrences of each glitch class (target column)
class_counts = df['ml_label'].value_counts()

# Sort for visual consistency
class_counts = class_counts.sort_values(ascending=False)

# --- Plot ---
plt.figure(figsize=(10, 5))
bars = plt.bar(class_counts.index, class_counts.values, color="steelblue", edgecolor="black")

# Log scale for y-axis
plt.yscale("log")

# Axis labels and title
plt.xlabel("Glitch Class (ml_label)", fontsize=11)
plt.ylabel("Count (log scale)", fontsize=11)
plt.title("Class Distribution of Glitch Types (log scale)", fontsize=13, pad=10)

# Rotate x-tick labels for readability
plt.xticks(rotation=45, ha="right", fontsize=9)

# Grid for minor lines
plt.grid(axis='y', linestyle='--', linewidth=0.5, alpha=0.7)

# Tight layout for clean spacing
plt.tight_layout()

# Save plot
plt.savefig("class_distribution_logscale.png", dpi=400, bbox_inches="tight")

plt.show()
```

```python
# Figure — PRD-optimized class distribution (log scale)

import pandas as pd
import matplotlib.pyplot as plt

# === Load data ===
df = pd.read_csv("original_data_cleaned.csv")

# === Compute class counts ===
class_counts = df['ml_label'].value_counts()
class_counts = class_counts.sort_values(ascending=False)

# === Global font settings (CRITICAL) ===
plt.rcParams.update({
    "font.size": 16,
    "axes.labelsize": 18,
    "axes.titlesize": 20,
    "xtick.labelsize": 14,
    "ytick.labelsize": 14
})

# === Plot ===
plt.figure(figsize=(12, 7))  # taller and wider for readability

bars = plt.bar(
    class_counts.index,
    class_counts.values,
    color="steelblue",
    edgecolor="black",
    linewidth=1.0
)

# === Log scale ===
plt.yscale("log")

# === Labels ===
plt.xlabel("Glitch class")
plt.ylabel("Count (log scale)")
plt.title("Class distribution of glitch types")

# === X ticks ===
plt.xticks(rotation=45, ha="right")

# === Grid ===
plt.grid(axis='y', linestyle='--', linewidth=0.6, alpha=0.4)

# === Layout ===
plt.tight_layout()

# === Save ===
plt.savefig(
    "class_distribution_logscale_PRD.png",
    dpi=600,
    bbox_inches="tight"
)

plt.savefig(
    "class_distribution_logscale_PRD.pdf",
    bbox_inches="tight"
)

plt.show()
```
